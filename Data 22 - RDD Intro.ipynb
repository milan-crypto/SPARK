```
import random

keyrdd = rdd.map(lambda x: (random.choice(['a', 'b', 'c']), x))
keyrdd.take(5)
```

```
keyrdd.take(5)
```

```
keyrdd.countByKey()
```

```
keyrdd.countByValue()
```

```
keyrdd.collectAsMap()
```

```
keyrdd.reduceByKey(lambda x, y: x + y).collect()
```

```
rdd.reduce(lambda x, y: x + y)
```

```
rdd.filter(lambda x: x != 0).reduce(lambda x, y: x * y)
```

```
keyrdd.groupByKey().map(lambda x: (x[0], list(x[1]))).collect()
```

```
keyrdd.groupByKey().mapValues(list).collect()
```

```
keyrdd = keyrdd.mapValues(lambda x: random.choice(range(1,101)))
keyrdd.take(5)
```

```
keyrdd.reduceByKey(lambda x, y: x + y).collect()
```

```
keyrdd.reduceByKey(lambda x, y: ((x + y) / y)).collect()
```

```
keyrdd.mapValues(lambda x: (x, 1))\
.reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))\
.mapValues(lambda x: x[0] / x[1])\
.collect()
```

```
header = iris.first()
iris_data = iris_raw.filter(lambda x: x != header)
iris.take(5)
```
